# Introduction

The Iris Dataset is one of the most famous datasets in machine learning and statistics. It was introduced by Ronald Fisher in 1936 and contains measurements of iris flowers from three different species.

# Problem Statement

The goal of this task is to:

Load and inspect the dataset

Understand its structure

Perform basic data exploration

Visualize relationships between features

Detect patterns and possible outliers

We will use:

pandas for data handling

matplotlib and seaborn for visualization

# Dataset Understanding and Description

The dataset contains 150 samples with 5 columns:

Feature	Description:

sepal_length:             	Length of sepal (cm)

sepal_width:                Width of sepal (cm)

petal_length:	            Length of petal (cm)

petal_width:             	Width of petal (cm)

species :	Type of iris flower

There are 3 species:

Setosa

Versicolor

Virginica

# Data Loading and Inspection

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn
df = sns.load_dataset("iris")

# Display dataset shape
print("Dataset Shape:", df.shape)

# Display column names
print("Columns:", df.columns)

# Display first 5 rows
df.head()
```

# Data Cleaning and Preparation

```python
# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# Check for duplicate values
print("Duplicate Rows:", df.duplicated().sum())
```

# Exploratory Data Analysis (EDA)

```python

#Scatter Plot  Relationship Between Features

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python

# Histogram  Distribution of Sepal Length

plt.figure(figsize=(8,6))
sns.histplot(df["sepal_length"], bins=20, kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.show()
```

```python

# Box Plot  Detecting Outliers

plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="species", y="petal_width")
plt.title("Petal Width Distribution by Species")
plt.show()
```

# Model Training and Testing (Optional Basic Classification)

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Separate features and target
X = df.drop("species", axis=1)
y = df["species"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
```

# Evaluation Metrics

```python
# Accuracy

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification Report

print("Classification Report:\n", classification_report(y_test, y_pred))
```
